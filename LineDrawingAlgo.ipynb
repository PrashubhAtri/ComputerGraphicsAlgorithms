```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('Solarize_Light2')
```

```python
import math
def apprx(n, decimals=0):
    multiplier = 10 ** decimals
    return math.floor(n*multiplier + 0.5) / multiplier
```

```python
def plot(points):
    plt.plot(points[0], points[1])
    plt.show()
```

DDA Line drawing

```python
def DDA_Line(start, end):
    xi,yi = start
    xf, yf = end
    x = xf - xi
    y = yf - yi
    m = y/x
    if(abs(x)>abs(y)):
        steps = x
    else:
        steps = y
    count = 1
    start = start.reshape(2,1)
    while ((xi==xf and yi==yf)==False and count<steps):
        if(m<1):
            xi = 1+xi
            yi = m+yi
        elif(m==1):
            xi = 1+xi
            yi = 1+yi
        else:
            xi = (1/m)+xi
            yi = 1+yi
        count+=1
        start = np.append(start,np.array([apprx(xi),apprx(yi)]).reshape(2,1), axis=1)
    start = np.append(start,end.reshape(2,1), axis=1)
    plot(start)
    return start
```

```python
start = np.array([5,6])
end = np.array([8,12])
res = DDA_Line(start, end)
print(res)
```

```python
start = np.array([5,6])
end = np.array([13,10])
res = DDA_Line(start, end)
print(res)
```

```python
start = np.array([1,7])
end = np.array([11,17])
res = DDA_Line(start, end)
print(res)
```

Bresenham's Line Algo

```python
def BLA(start, end):
    xi,yi = start
    xf, yf = end
    x = xf - xi
    y = yf - yi
    p = 2*y - x
    itr = 0
    start = start.reshape(2,1)
    while((xi==xf and yi==yf)==False and itr!=(x-1)):
        if(p<0):
            p = p+2*y
            xi = xi+1
            yi=yi
        else:
            p = p + 2*y - 2*x
            xi = xi+1
            yi = yi+1
        itr+=1
        start = np.append(start,np.array([xi,yi]).reshape(2,1), axis=1)
    start = np.append(start,end.reshape(2,1), axis=1)
    plot(start)
    return start
```

```python
start = np.array([9,18])
end = np.array([14,22])
res = DDA_Line(start, end)
print(res)
res = BLA(start, end)
print(res)
```

```python
start = np.array([20,10])
end = np.array([30,18])
res = DDA_Line(start, end)
print(res)
res = BLA(start, end)
print(res)
```

Mid Point Line Drawing Algo

```python
def MPLDA(start, end):
    xi,yi = start
    xf, yf = end
    x = xf - xi
    y = yf - yi
    Di = 2*y - x
    D = 2*(y - x)
    itr = 0
    start = start.reshape(2,1)
    while((xi==xf and yi==yf)==False and itr!=(x-1)):
        if(Di<0):
            Di = Di + 2*y
            xi = xi + 1
            yi = yi
        else:
            Di = Di + D
            xi = xi + 1
            yi = yi + 1
        itr+=1
        start = np.append(start,np.array([xi,yi]).reshape(2,1), axis=1)
    start = np.append(start,end.reshape(2,1), axis=1)
    plot(start)
    return start
```

```python
start = np.array([20,10])
end = np.array([30,18])
res = DDA_Line(start, end)
print(res)
res = BLA(start, end)
print(res)
res = MPLDA(start, end)
print(res)
```

```python
start = np.array([5,9])
end = np.array([12,16])
res = DDA_Line(start, end)
print(res)
res = BLA(start, end)
print(res)
res = MPLDA(start, end)
print(res)
```
